```python
# 이미지 경로 설정
image_path = '.../low_tissue_15360_14336.png'
```

```python
'''
첫 번째 고유벡터를 법선 벡터로 하는 평면에 대해 사영
'''
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt

# 이미지 로드 및 RGB로 변환
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 이미지 데이터를 (픽셀 수, 채널 수) 형태로 변환
pixels = image.reshape(-1, 3)

# 데이터 평균 계산 및 중심화
X_mean = np.mean(pixels, axis=0)
X_centered = pixels - X_mean

# 공분산 행렬 계산
cov_matrix = np.cov(X_centered, rowvar=False)

# 고유값 및 고유벡터 계산
eigvals, eigvecs = np.linalg.eig(cov_matrix)

# 고유값을 내림차순으로 정렬 (큰 값부터)
sort_indices = np.argsort(eigvals)[::-1]
sort_eigvals = eigvals[sort_indices]
sort_eigvecs = eigvecs[:, sort_indices]

# 두 번째와 세 번째 고유벡터 선택
second_eigenvector = sort_eigvecs[:, 1]
third_eigenvector = sort_eigvecs[:, 2]

# 데이터 사영 (두 번째와 세 번째 고유벡터 방향으로)
projection_matrix = np.column_stack((second_eigenvector, third_eigenvector))
projected_data_2d = X_centered.dot(projection_matrix)

# 결과 시각화 (2차원 평면에서)
plt.figure(figsize=(8, 6))
plt.scatter(projected_data_2d[:, 0], projected_data_2d[:, 1],c = pixels/255, s=1, alpha=0.2)
plt.title('Projection onto Plane Defined by Second and Third Eigenvectors')
plt.xlabel('Second Eigenvector')
plt.ylabel('Third Eigenvector')
plt.grid(True)
# plt.savefig('/home/dororo99/prof_task/analyze/marking/20580_10632/projection.png')
```
